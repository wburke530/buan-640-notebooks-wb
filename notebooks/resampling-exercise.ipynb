Now, we randomly select five observations and compute the mean. We repeat this experiment 1000 times:

```python
sample_means_5 = []

for i in range(1000):
    sample = homes["age"].sample(5)
    sample_means_5.append(sample.mean())
plt.hist(sample_means_5, bins=20)
plt.xlabel("Sample Mean Age")
plt.ylabel("Frequency")
plt.title("Distribution of Sample Means (n = 5)")
plt.show()
```

We increase the sample size to 60 and repeat the experiment. What do you observe?
